Unigram Model

```python
import nltk
from nltk import FreqDist
from nltk.tokenize import WhitespaceTokenizer
corpus = "<s> I am Sam </s> <s> Sam I am </s> <s> I do not like green eggs and ham </s>" #corpus 
token_obj = WhitespaceTokenizer() #tokenizer obj to tokenize the coupus on the bases of whitespaces
tokens = token_obj.tokenize(corpus) #tokenizing
print("all tokens in corpus : ",tokens)

total_words = len(tokens) #corpus size
print("corpus size : ",total_words)

vocabulary = list(set(tokens))  #vocabulary 
print("vocabulary : ",vocabulary)  

vocabulary_size = len(vocabulary) #vocabulary size
print("vocabulary size : ",vocabulary_size)


###################  part 1  #################################
unigram_counts = {}
unigram_model = {}
for i in range(0,vocabulary_size):#to find the count of each unique word
    count=0
    for j in range(0,total_words):
        if(vocabulary[i]==tokens[j]):
            count=count+1
    unigram_counts[vocabulary[i]] = count        #dictionary with unique counts
    unigram_model[vocabulary[i]] = count/total_words  # dictionary holding the unigram model

print("unigram counts: ",unigram_counts)
print("single value",unigram_counts['<s>'])
print("unigram model: ",unigram_model)
print("single value",unigram_model['<s>'])
print("\n\n")
```

Bigram Model

```python
corpus = "<s> I am Sam </s> <s> Sam I am </s> <s> I do not like green eggs and ham </s>" #corpus 
token_obj = WhitespaceTokenizer() #tokenizer obj to tokenize the coupus on the bases of whitespaces
tokens = token_obj.tokenize(corpus) #tokenizing
print("all tokens in corpus : ",tokens)

total_words = len(tokens) #corpus size
print("corpus size : ",total_words)

vocabulary = list(set(tokens))  #vocabulary 
print("vocabulary : ",vocabulary)  

vocabulary_size = len(vocabulary) #vocabulary size
print("vocabulary size : ",vocabulary_size)

###################  part 2  #################################
bigram_tokens = []
for i in range(0,vocabulary_size):#to find the count of each unique word
    for j in range(0,vocabulary_size):
        bigram_tokens.append((vocabulary[i],vocabulary[j]))

bigram_counts = {}
bigram_model = {}
for i in range(0,len(bigram_tokens)):#to find the count of each unique word
    count=0
    for j in range(0,total_words-1):
        if((tokens[j],tokens[j+1])==bigram_tokens[i]):
            count=count+1
    bigram_counts[bigram_tokens[i]] = count        #dictionary with unique counts
    bigram_model[bigram_tokens[i]] = (count+1)/(unigram_counts[bigram_tokens[i][0]] +vocabulary_size) # dictionary holding the bigram model
    
# for pair in bigram_tokens:
#     print(pair)
print("bigram counts: ",bigram_counts)
print("single value",bigram_counts[('<s>', 'I')])
print("bigram model: ",bigram_model)
print("single value",bigram_model[('<s>', 'I')])
print("\n\n")
```
