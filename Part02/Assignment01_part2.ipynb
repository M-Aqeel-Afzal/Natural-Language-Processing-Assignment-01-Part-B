```python
import numpy as np
import random
# defining the dataset
dataset = [['This book is good, very good', '+'],
              ['This book is bad, very bad', '-'],
              ['This novel is good', '+'],
              ['This novel is bad', '-'],
              ['This book is some good and some bad', '0'],
              ['This novel is little good but little bad', '0']]

classes = ['Good', 'Bad', 'neutral'] # classes
# getting vocabulary
single_string=""
for sentence in dataset:
  single_string+=single_string+sentence[0]+" "
single_string = single_string.replace(",", "")
print(single_string)
words = single_string.lower().split()
vocabulary = set(words)
print(vocabulary)


dict_negative_count = {}
dict_positive_count = {}
dict_neutral_count = {}
def dict_count():   # function for getting each word count in each class
  for word in vocabulary:
    dict_positive_count[word]=0
    dict_negative_count[word]=0
    dict_neutral_count[word]=0

  for sentence in dataset:
    single_string=sentence[0]
    single_string = single_string.replace(",", "")
    words = single_string.lower().split()
    if sentence[1]=='+':
      for word in words:
        dict_positive_count[word]+=1
    if sentence[1]=='-':
      for word in words:
        dict_negative_count[word]+=1
    if sentence[1]=='0':
      for word in words:
        dict_neutral_count[word]+=1

  print("positve each word counts:", dict_positive_count)
  print("negative each word counts:", dict_negative_count)
  print("neutral each word counts:", dict_neutral_count)

# P(word|class) = (count of word in class + 1) / (total count of words in class + total count of unique words)


dict_negative_conditional = {}
dict_positive_conditional = {}
dict_neutral_conditional = {}
def conditional_prob():     # function for getting each word conditional probability
  total_positive_count=0
  total_negative_count=0
  total_neutral_count=0
  for word in vocabulary:
    dict_positive_conditional[word]=0
    dict_negative_conditional[word]=0
    dict_neutral_conditional[word]=0
    total_positive_count +=dict_positive_count[word]
    total_negative_count +=dict_negative_count[word]
    total_neutral_count +=dict_neutral_count[word]

  for word in vocabulary:
    dict_positive_conditional[word] +=(dict_positive_count[word]+1)/(total_positive_count+len(vocabulary))
    dict_negative_conditional[word] +=(dict_negative_count[word]+1)/(total_negative_count+len(vocabulary))
    dict_neutral_conditional[word] +=(dict_neutral_count[word]+1)/(total_neutral_count+len(vocabulary))

  print("positve class total count:", total_positive_count)
  print("negative class total count:", total_negative_count)
  print("neutral class total count:", total_neutral_count)

  print("positve conditional prob:", dict_positive_conditional)
  print("negative conditional prob:", dict_negative_conditional)
  print("neutral conditional prob:", dict_neutral_conditional)

prior_positve = 0
prior_negative = 0
prior_neutral = 0
def prior_prob(): # function for getting prior probability
  global prior_positve
  global prior_negative
  global prior_neutral
  for doc in dataset:
    if doc[1]=='+':
      prior_positve+=1
    if doc[1]=='-':
      prior_negative+=1
    if doc[1]=='0':
      prior_neutral+=1

  prior_positve = prior_positve/len(dataset)
  prior_negative = prior_negative/len(dataset)
  prior_neutral = prior_neutral/len(dataset)

  print("positve prior prob:", prior_positve)
  print("negative prior prob:", prior_negative)
  print("neutral prior prob:", prior_neutral)




def predictor(new_word):   # function to pridict the class of a sentence
  new_word = new_word.replace(",", "")
  new_word = new_word.replace(".", "")
  words = new_word.lower().split()
  positive_prob = 1
  negative_prob = 1
  neutral_prob = 1
  for word in words:
    positive_prob *= dict_positive_conditional[word]
    negative_prob *= dict_negative_conditional[word]
    neutral_prob *= dict_neutral_conditional[word]

  p=[]
  p.append(positive_prob)
  p.append(negative_prob)
  p.append(neutral_prob)   
  return  np.argmax(p),p


def genrate():
  print("\n\n\n")
  global vocabulary
  vocabulary = list(vocabulary)
  sentence=""
  flag1=False
  flag2=False
  flag3=False
  index = []
  p = []
  while(not(flag1 and flag2 and flag3)):
    for i in range(0,4):
      sentence+=random.choice(vocabulary)+" "

    index,p=predictor(sentence)
    if(index==0 and not flag1):
      print("probilies:", p)
      print("\" ",sentence," \" belongs to:   ", classes[index])
      flag1=True
    if(index==1 and not flag2):
      print("probilies:", p)
      print("\" ",sentence," \" belongs to:   ", classes[index])
      flag2=True
    if(index==2 and not flag3):
      print("probilies:", p)
      print("\" ",sentence," \" belongs to:   ", classes[index])
      flag3=True
    sentence=""



new_word = 'This book is good'
dict_count()
conditional_prob()
prior_prob()
index,p=predictor(new_word)
print("probilies:", p)
print("\" ",new_word," \" belongs to:   ", classes[index])
genrate()
```
